```python
import os
from glob import glob

rootdir = './sites'
sub_dirs = glob(os.path.join(rootdir, '*'))

nouns_list = []
label_list = []

for i, sub in enumerate(sub_dirs):
    print(i, sub)
    files = glob(os.path.join(sub, '*.noun'))
    for filename in files:
        with open(filename, 'r') as f:
            words = f.read().split(' ')
            nouns_list.append(words)
            label_list.append(i)    
```

```python
from gensim import corpora, matutils
```

```python
dictionary = corpora.Dictionary(nouns_list)
len(dictionary)
```

```python
dictionary.save_as_text('sakuradic.txt')
```

```python
dictionary = corpora.Dictionary.load_from_text('sakuradic.txt')
len(dictionary)
```

```python
dictionary.filter_extremes(no_above=0.5, no_below=2)
len(dictionary)
```

```python
n_words = len(dictionary)

# Make bags of words
bows_list = []
for nouns in nouns_list:
    bow_id = dictionary.doc2bow(nouns)
    bow = matutils.corpus2dense([bow_id], n_words).T[0]
    bows_list.append(bow)
```

```python
from sklearn.model_selection import train_test_split

data_train, data_test, label_train, label_test = train_test_split(bows_list, label_list, test_size=0.1)
```

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier()
classifier.fit(data_train, label_train)
```

```python
print(classifier.score(data_test, label_test))
```

```python
from sklearn.metrics import classification_report

y_true = label_test
y_pred = classifier.predict(data_test)

print(classification_report(y_true, y_pred))
```
